```python
import os
import pandas as pd
import lasio

def find_closest_depth(las_depth, csv_depths):
    closest_depth = csv_depths.iloc[(csv_depths - las_depth).abs().argsort()[:1]]
    return closest_depth.values[0]
```

```python
# 204-19-7
las_path = 'Sources/data/wirelineLog/204-19-7_logs.las'
csv_path = 'Sources/data/labels/permeability/204_19_7.csv'  
file_path = f'Sources/data/labels/permeability/preprocessed_data'
las_data = lasio.read(las_path).df()
csv_data = pd.read_csv(csv_path)
csv_data = csv_data.dropna(subset=['DEPTH'])
depth = list(csv_data['DEPTH'])
for i, value in enumerate(depth):
    if not isinstance(value, float):
        result = ''.join(['0' if c.isalpha() else c for c in value])
        result = result.replace(',','.')
        depth[i] = float(result)
csv_data['DEPTH'] = depth

csv_data.loc[:, 'DEPTH'] = csv_data['DEPTH'].astype(float)

csv_data.loc[(csv_data['DEPTH'] >= 2131.00) & (csv_data['DEPTH'] < 2157.68), 'DEPTH'] += 1.8

csv_data.loc[(csv_data['DEPTH'] >= 2158.00) & (csv_data['DEPTH'] < 2180.38), 'DEPTH'] += 1.6

csv_data.loc[(csv_data['DEPTH'] >= 2540.00) & (csv_data['DEPTH'] < 2544.89), 'DEPTH'] += 3.0

csv_data = csv_data.dropna(subset = ['PERMEABILITY (HORIZONTAL)\nKair\nmd'])
las_data = las_data[(las_data.index >= csv_data['DEPTH'].min()) & (las_data.index <= csv_data['DEPTH'].max())]
las_data['closest_depth'] = las_data.index.to_series().apply(lambda d: find_closest_depth(d, csv_data['DEPTH']))
las_data['true depth'] = las_data.index
csv_data.set_index('DEPTH', inplace=True)
merged_data = pd.merge(las_data, csv_data, left_on='closest_depth', right_index=True, how='left')
diff = abs(merged_data['closest_depth'] - merged_data['true depth'])
remove_idx = diff[diff>0.5].index
merged_data = merged_data.drop(index = remove_idx)
merged_data.to_csv(file_path+'/204_19_7(preprocessed).csv', index=False)
```
